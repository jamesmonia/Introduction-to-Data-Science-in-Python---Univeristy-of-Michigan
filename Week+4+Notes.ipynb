```python
chance = 0.01/100
np.random.binomial(10000,chance)
#runs the sim as pass or fail once by default
```

```python
tornado = np.random.binomial(1,chance,1000000)
```

```python
chance = 0.01
two_days_in_a_row = 0 
for j in range(1,len(tornado)-1):
    if tornado[j] == 1 and tornado[j-1] == 1:
        two_days_in_a_row += 1
        
print(f'Two days in a row is {two_days_in_a_row}')
```
